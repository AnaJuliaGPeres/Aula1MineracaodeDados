>Essa linha de código abaixo cria um dicionário em Python que associa um número (chave) a um nome de mês (valor).O dicionário permite uma associação rápida e flexível entre as informações, e pode ser muito útil em diversas situações em que precisamos mapear valores para outras informações.Cada chave é um número inteiro que representa um mês e cada valor é uma string que representa o nome do mês correspondente.

```python
#criando um dicionário
meses={1: 'Janeiro',2: 'Fevereiro', 3: 'Março'}
```

>A função dt.month é um método do Pandas que é aplicado a um objeto do tipo datetime e retorna uma série com o mês correspondente a cada data.A função map é utilizada em seguida para mapear o número do mês para o seu respectivo nome.Ela acaba aplicando a função especificada a cada elemento da série, ou seja, a cada número de mês extraído da coluna 'Data' após a conversão para um objeto de data e hora (datetime). Neste caso, a função especificada é meses, que é o dicionário que mapeia cada número de mês para seu nome correspondente. Dessa forma, esse código abaixo cria uma nova coluna chamada 'Data' no DataFrame df, que contém os nomes dos meses correspondentes aos valores da coluna 'Data' original.

```python
#mapear o número mês para string mês
df['Data']=pd.to_datetime(df['Data']).dt.month.map(meses)
```

```python
df
```

>Fazendo a transformação do dataframe  somente no horário em object(String).
>Para isso utilizamos a função dt.strftime() que é um método usado para formatar uma data em uma string. A função recebe como argumento uma string de formatação, que define como a data deve ser apresentada na forma de uma string.A string de formatação '%H' é usada para extrair somente a hora de um objeto do tipo datetime. A letra 'H' representa a hora em formato de 24 horas, e % é um caractere especial que indica que um código de formatação segue. O resultado é então atribuído à mesma coluna, sobrescrevendo os valores originais.Isso permite que ela seja   mais facilmente manipulada ou apresentada.

```python
#transformar datatime em object (string), selecionado apenas o horario
df['Horário']=df['Horário'].dt.strftime('%H')
```

```python
df
```

> Modificando na coluna 'Horário' as horas por Manhã,tarde e noite.

> Para isso foi utilizado um laço de repetição que inicia-se como uma lista vazia e posteriormente vai sendo preenchendo com o script. Se o número que está no  campo Horário for maior que 8 e menor que 13 então vai ser definido como Manhã. Se o número for maior que 12 e menor que 19 então vai ser Tarde e por fim se não é nenhum desses fica definido como Noite. 

>No final foi passado para o dataframe que [horario] seria igual a time, que é a lista criada com as condições;


```python
time = []#lista vazia
for i in list(map(int,df['Horário'])):#laço for mapeando e transformando string para inteiro
   if i > 8 and i < 13:#9-12(manha)
     time.append('Manhã')#comando append vai inserir manha quando o criterio for verdadeiro na lista time
   elif i > 12 and i < 19:
      time.append('Tarde')
   else:
      time.append('Noite')
df['Horário'] = time #substituindo a coluna horario pela lista time
```

### 1 Análise

>Para começar com a análise de dados começamos agrupando linha de produtos com a avaliação. Dentro da função groupby(), 'Linha de produto' é passado como uma string entre parênteses, indicando que é o nome da coluna que será usada para agrupar os dados. Já 'Avaliação' é passado como uma string dentro de colchetes, indicando que é a coluna cujos valores serão agregados dentro de cada grupo criado pela função groupby(). O 'mean' calcula a média de valores de uma coluna. O reset_index(drop=False) reorganiza o índice do DataFrame após a operação de agrupamento. E o drop=False indica que a coluna usada para agrupamento (Linha de produto) não deve ser descartada do resultado final.

```python
agrupamento = df.groupby('Linha de produto')['Avaliação'].mean().reset_index(drop=False)
```

>Mostrando a tabela do agrupamento realizado. 

```python
agrupamento
```

> Arredondando o valor da avaliação para apenas uma casa decimal após a virgula.

```python
#Arredondando para uma casa decimal
agrupamento_avaliaçao = agrupamento.round(1)
```

>Mostrando o resultado do agrupamento com o arredondamento realizado.

```python
agrupamento_avaliaçao
```

>Pegando o agrupamento realizado das avaliações com a linha de produtos e colocando-os em ordem decrescente. O comando ascending é usado para especificar a ordem da classificação, False indica que os valores devem ser classificados em ordem decrescente.

```python
#classificando valores na ordem decrescente
agrupamento_avaliaçao = agrupamento_avaliaçao.sort_values(by = 'Avaliação', ascending = False)
```

> Apartir dos dados agrupados criamos uma forma de visualiza-los. Foi utilizado a biblioteca gnuplot para nos ajuda a plotar os gráficos.Por fim foi invertido a sequência de cores para ter uma melhor visualização.

> Para fazermos o gráfico de cores mostrando a relação entre a linha de produtos com a avaliações usamos o barplot para desenhar o gráfico, para assim, visualizarmos quais produtos foram melhor avaliados.

```python
ax = sns.barplot(y = 'Linha de produto', x = 'Avaliação', data = agrupamento_avaliaçao, palette = Figura)
```

> Foi criado um novo agrupamento, dessa vez 'linha de produto' e 'lucro'. É aplicado o método sum() na coluna 'Lucro', que é a coluna que se deseja somar para cada grupo de valores de 'Linha de produto'. O parâmetro reset_index() foi usado para resetar os índices do DataFrame resultante, deixando o índice padrão. E o drop=False mantem a coluna 'Linha de produto' como uma coluna do DataFrame resultante.

```python
Agrupamento_lucro_linhadeproduto= df.groupby('Linha de produto')['Lucro'].sum().reset_index(drop = False)
```

> Apartir do agrupamento realizado, foi criado um gráfico em barras, utilizando gnuplot para selecionar as cores do gráfico e depois invertendo-as. Então, é criado um gráfico de barras usando o método barplot e passado como parâmetro a coluna 'Linha de produto' para o eixo x e a coluna 'Lucro' para o eixo y.Logo utilizamos o ax.get_xticklabels que retorna os valores do eixo x e passamos para ele novas configurações personalizadas.O fontsize=12 define o tamanho das fontes e o rotation=45 define a inclinação das letras. O ha='right' alinha os rótulos mais à direita.

```python
Agrupamento_lucro_linhadeproduto = Agrupamento_lucro_linhadeproduto.sort_values(by = 'Lucro', ascending = False)
Figura = sns.color_palette('gnuplot')#selecionando as cores do grafico
Figura.reverse()#invertendo a sequencia de cores
ax = sns.barplot(x = 'Linha de produto',y = 'Lucro', data = Agrupamento_lucro_linhadeproduto, palette = Figura)
ax.set_xticklabels(ax.get_xticklabels(), fontsize=12, rotation=45, ha='right')
```

> Pelo gráfico acima podemos perceber que a linha de produtos que mais dá lucro é a de saúde e beleza e acessórios eletrônicos é a que menos dá lucro.
